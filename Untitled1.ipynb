```python
X = X.drop(['workclass', 'education', 'race', 'sex','capital.loss', 'native.country'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
for col in X.columns:
  df[col] = ss.fit_transform(df[col].values.reshape(-1,1))
```

```python
round(y.value_counts(normalize=True) * 100, 2).astype('str') + ' %'
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
```

```python
ros.fit(X,y)
```

```python
X_resampled,y_resampled = ros.fit_resample(X,y)
```

```python
#round(y_resampled.value_counts(normalize=True) * 100, 2).astype('str') + ' %'
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("Y_train shape:", y_train.shape)
print("Y_test shape:", y_test.shape)
```

```python
from sklearn.metrics import accuracy_score,f1_score
def generate_scores(y2, y1):
    print("Accuracy = ", accuracy_score(y2, y1))
    print("F1 = ", f1_score(y2, y1))
    pass
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression(random_state =43)
lg.fit(X_train,y_train)
y_pred_lg = lg.predict(X_test)
generate_scores(y_test,y_pred_lg)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier()
knc.fit(X_train,y_train)
y_pred_knc = knc.predict(X_test)
generate_scores(y_test,y_pred_knc)
```

```python
from sklearn.naive_bayes import GaussianNB
gb = GaussianNB()
gb.fit(X_train,y_train)
y_pred_gb = gb.predict(X_test)
generate_scores(y_test,y_pred_gb)
```

```python
from sklearn.svm import SVC
svc = SVC(random_state =43)
svc.fit(X_train,y_train)
y_pred_svc = svc.predict(X_test)
generate_scores(y_test,y_pred_svc)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state =43)
dtc.fit(X_train,y_train)
y_pred_dtc = dtc.predict(X_test)
generate_scores(y_test,y_pred_dtc)
```

```python
from sklearn.ensemble import RandomForestClassifier
rdc = RandomForestClassifier(random_state =43)
rdc.fit(X_train,y_train)
y_pred_rdc = rdc.predict(X_test)
generate_scores(y_test,y_pred_rdc)
```

```python
from xgboost import  XGBClassifier
xgbc = XGBClassifier(random_state =43)
xgbc.fit(X_train,y_train)
y_pred_xgbc = xgbc.predict(X_test)
generate_scores(y_test,y_pred_xgbc)
```

```python
from scipy.stats import randint
```

```python
r_tune = RandomForestClassifier(random_state = 43,n_jobs =-1)
```

```python
from sklearn.model_selection import RandomizedSearchCV
rc_dist = {'max_depth' : np.arange(40,150,15),
           'n_estimators' : np.arange(40,150,15),
           'max_features': randint(1,3),
           'criterion': ['gini','entropy'],
           'bootstrap': [True,False],
           'min_samples_leaf': randint(1,4)}
```

```python
def hypertuningrf_cv(est,p_dist,nb_iter,X,y):
  rdsearch = RandomizedSearchCV(est,param_distributions=rc_dist,n_jobs =-1,n_iter = nb_iter,cv =5)
  rdsearch.fit(X,y)
  ht_params = rdsearch.best_params_
  ht_score = rdsearch.best_score_
  return ht_params,ht_score
```

```python
rf_parameters,rf_ht_score = hypertuningrf_cv(r_tune,rc_dist,40,X_train,y_train)
```

```python
 print('rf_parameters :',rf_parameters)
 print('rf_ht_score:',rf_ht_score)
```

```python
rf_final = RandomForestClassifier(bootstrap=False,criterion='gini',max_depth=130,max_features=2,min_samples_leaf=1,n_estimators=145)
rf_final.fit(X_train,y_train)
y_pred_rf_final = rf_final.predict(X_test)
generate_scores(y_test,y_pred_rf_final)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred_rf_final)
cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_rf_final))
```
